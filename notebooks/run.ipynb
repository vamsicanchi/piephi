```python
import os
from pdf2image import convert_from_path
from PIL import Image
from PyPDF2 import PdfWriter

# Function to compress an image
def compress_image(input_image, output_image, quality=15):
    image = Image.open(input_image)
    # Save the image with reduced quality
    image.save(output_image, quality=quality, optimize=True)

# Function to convert PDF to images and compress them
def pdf_to_images_and_compress(pdf_path, output_folder):
    # Convert PDF pages to images
    images = convert_from_path(pdf_path, poppler_path=r"D:\programs\poppler-23.11.0\Library\bin")

    # Create a directory to store the images
    os.makedirs(output_folder, exist_ok=True)

    compressed_image_paths = []

    # Save images and compress them
    for i, image in enumerate(images):
        image_path = os.path.join(output_folder, f"page_{i + 1}.jpg")
        compressed_image_path = os.path.join(output_folder, f"compressed_page_{i + 1}.jpg")

        # Save the original image
        image.save(image_path)

        # Compress the image
        compress_image(image_path, compressed_image_path)

        compressed_image_paths.append(compressed_image_path)

    return compressed_image_paths

# Function to convert PDF to images and compress them
def pdf_merger(pdf_path1, pdf_path2, output_folder):
    # Convert PDF pages to images
    images1 = convert_from_path(pdf_path1, poppler_path=r"D:\programs\poppler-23.11.0\Library\bin")

    images2 = convert_from_path(pdf_path2, poppler_path=r"D:\programs\poppler-23.11.0\Library\bin")

    # Create a directory to store the images
    os.makedirs(output_folder, exist_ok=True)

    images = images1+images2

    images_path = []

    # Save images
    for i, image in enumerate(images):
        image_path = os.path.join(output_folder, f"page_{i + 1}.jpg")
        images_path.append(image_path)
        image.save(image_path)

    return images_path


def images_to_pdf(image_paths, output_pdf):

    images = [Image.open(f) for f in image_paths]

    images[0].save(output_pdf, "PDF" ,resolution=100.0, save_all=True, append_images=images[1:])



# # Paths
# input_pdf_path = 'D:\downloads\petrolbunk\sharmila\leasedeed.pdf'  # Replace with your input PDF path
# output_folder_path = 'D:\downloads\petrolbunk\sharmila\leasedeed'  # Folder to store converted and compressed images
# output_pdf_path = 'D:\downloads\petrolbunk\sharmila\leasedeed_compressed.pdf'  # Path to save the compressed PDF

# # Convert PDF to images and compress them
# compressed_images = pdf_to_images_and_compress(input_pdf_path, output_folder_path)

# # Create a PDF from compressed images
# images_to_pdf(compressed_images, output_pdf_path)

# print(f"PDF created: {output_pdf_path}")

pdf_path1 = "D:\downloads\petrolbunk\sharmila\khasra.pdf"
pdf_path2 = "D:\downloads\petrolbunk\sharmila\saledeed.pdf"
output_folder = 'D:\downloads\petrolbunk\sharmila\khasrasaledeed'
pdf_merged = "D:\downloads\petrolbunk\sharmila\khasra_saledeed.pdf"

images = pdf_merger(pdf_path1, pdf_path2, output_folder)


images_to_pdf(images, pdf_merged)
```
