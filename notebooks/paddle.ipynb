```python
import os
import cv2
os.chdir("D:\code\python\piephi")
from library.ocr import ocr
```

```python
from appconfig import properties
```

```python
img_path = r'D:\code\python\piephi\samples\images\sample3.png'
image = cv2.imread(img_path)
pocr_obj = ocr.Paddle(properties["paddle"], "log")
result = pocr_obj.paddle_run(image)
for idx in range(len(result)):
    res = result[idx]
    for line in res:
        print(line)
```

```python
from paddleocr import draw_ocr
# draw result
from PIL import Image
result = result[0]
image = Image.open(img_path).convert('RGB')
boxes = [line[0] for line in result]
txts = [line[1][0] for line in result]
scores = [line[1][1] for line in result]
im_show = draw_ocr(image, boxes, txts, scores, font_path=r'D:\code\python\piephi\bin\models\paddledata\simfang.ttf')
im_show = Image.fromarray(im_show)
im_show.save('result1.jpg')
```
